```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset["rbc"]=le.fit_transform(dataset["rbc"])
dataset["pc"]=le.fit_transform(dataset["pc"])
dataset["pcc"]=le.fit_transform(dataset["pcc"])
dataset["ba"]=le.fit_transform(dataset["ba"])
dataset["pcv"]=le.fit_transform(dataset["pcv"])
dataset["wc"]=le.fit_transform(dataset["wc"])
dataset["rc"]=le.fit_transform(dataset["rc"])
dataset["htn"]=le.fit_transform(dataset["htn"])
dataset["dm"]=le.fit_transform(dataset["dm"])
dataset["cad"]=le.fit_transform(dataset["cad"])
dataset["appet"]=le.fit_transform(dataset["appet"])
dataset["pe"]=le.fit_transform(dataset["pe"])
dataset["ane"]=le.fit_transform(dataset["ane"])
dataset["classification"]=le.fit_transform(dataset["classification"])
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
y
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
models =[]
models.append(("LR",LogisticRegression()))
models.append(("CART",DecisionTreeClassifier()))
models.append(("LDA",LinearDiscriminantAnalysis()))
models.append(("KNN",KNeighborsClassifier()))
models.append(("NB",GaussianNB()))
models.append(("svm",SVC()))
names =[]
results = []
for name,model in models:
    kfold = model_selection.ShuffleSplit(n_splits =10,test_size = 0.33,random_state = 9)
    cv_results = model_selection.cross_val_score(model,x,y,cv = kfold,scoring = "accuracy")
    results.append(cv_results)
    names.append(name)
    print("%s: %f (%f)"% (name,cv_results.mean()*100,cv_results.std()))
```

```python
fig = plt.figure()
fig.suptitle("Algorithms Comparision")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
pipelines = []
pipelines.append(("ScaledLR",Pipeline([("Scaler",StandardScaler()),("LR",LogisticRegression())])))
pipelines.append(("ScaledCART",Pipeline([("Scaler",StandardScaler()),("CART",DecisionTreeClassifier())])))
pipelines.append(("ScaledLDA",Pipeline([("Scaler",StandardScaler()),("LDA",LinearDiscriminantAnalysis())])))
pipelines.append(("ScaledKNN",Pipeline([("Scaler",StandardScaler()),("KNN",KNeighborsClassifier())])))
pipelines.append(("ScaledNB",Pipeline([("Scaler",StandardScaler()),("NB",GaussianNB())])))
pipelines.append(("Scaledsvm",Pipeline([("Scaler",StandardScaler()),("Svm",SVC())])))
results = []
names = []
for name,model in models:
    kfold = model_selection.ShuffleSplit(n_splits =12,test_size =0.33,random_state=10)
    cv_results = model_selection.cross_val_score(model,x,y,cv = kfold ,scoring = "accuracy")
    results.append(cv_results)
    names.append(name)
    print("%s: %f (%f)"%(name,cv_results.mean()*100,cv_results.std()))
```

```python
fig =plt.figure()
fig.suptitle("Algorithms Comparision")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
ensembles = []
ensembles.append(("ADC",AdaBoostClassifier()))
ensembles.append(("GBM",GradientBoostingClassifier()))
ensembles.append(("RFC",RandomForestClassifier()))
ensembles.append(("ET",ExtraTreesClassifier()))
results = []
names = []
for name,model in ensembles:
    kfold = model_selection.ShuffleSplit(n_splits = 10,test_size = 0.33,random_state = 9)
    cv_results = model_selection.cross_val_score(model,x,y,cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_results)
    print("%s: %f (%f)"% (name,cv_results.mean()*100,cv_results.std()))
```

```python
fig =plt.figure()
fig.suptitle("Ensembles Algorithms Comparision")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
scaler = StandardScaler().fit(x)
rescaledX = scaler.transform(x)
model = ExtraTreesClassifier()
model.fit(rescaledX,y)
```

```python
rescaledValidations = scaler.transform(x)
predictions = model.predict(rescaledValidations)
print(accuracy_score(predictions,y))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y,predictions))
```
